```python
import requests
import pandas as pd
```

```python
# url = 'http://api-pokemon-baron.herokuapp.com/pokemon?generation=1&legendary=True'
url = 'http://api-pokemon-baron.herokuapp.com/pokemon'
# url = 'http://api-pokemon-baron.herokuapp.com/pokemon/145'
```

```python
# res = requests.get(url)
res = requests.get(url, params={ "generation" : 1, "legendary" : "True"})
```

# Access Local JSON File

### 1st Example

```python
import json
```

```python
pokemonList = []

with open('1stlegendary.json') as json_file:  
    data = json.load(json_file)
   
    for item in data :
        pokemonList.append(item)
        
dfPokemon = pd.DataFrame(pokemonList, columns=pokemonList[0].keys())
dfPokemon
```

```python
pokemonList = []

with open('2ndlegendary.json') as json_file:  
    data = json.load(json_file)
   
    for item in data['data'] :
        pokemonList.append(item)
        
dfPokemon = pd.DataFrame(pokemonList, columns=data['columns'], index=data['index'])
dfPokemon
```

### 2nd Example

```python
dfPokemon = pd.read_json('1stlegendary.json', orient='records')
dfPokemon
```
